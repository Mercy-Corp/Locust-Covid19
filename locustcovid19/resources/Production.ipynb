```python
import pandas as pd
from datetime import date
```

```python
production_df = pd.read_csv("git/Locust-Covid19/my_package/data/input/FAOSTAT_data_6-30-2020.csv", sep = ",")
production_df.head()
```

```python
# Create measure df

# initialize list of lists 
measure_id_data = [[12, 'Maize'], [13, 'Rice, paddy'], [14, 'Millet'], [15, 'Beans, dry'], [16, 'Wheat']]

# Create the pandas DataFrame 
measure_df = pd.DataFrame(measure_id_data, columns = ['measureID', 'Item'])

measure_df.head()
```

```python
# Merge production with measure
production_df = production_df.merge(measure_df, on = 'Item', how = 'left')
production_df.head()
```

```python
# Add fact id
production_df['factID'] = 'Dem_' + production_df['Area Code'].astype(str) + '_' + production_df['Item Code'].astype(str) + '_' + production_df['Year Code'].astype(str)
production_df.head()
```

```python
# Add date id
#Transform year column to datetime
production_df['Year'] = pd.to_datetime([f'{y}-01-01' for y in production_df.Year])
production_df.head()
```

```python
# Import date table
dates = pd.read_csv("git/Locust-Covid19/my_package/data/input/date_23_06-2020.csv")
dates['date'] = pd.to_datetime(dates['date'])
dates.head()
```

```python
production_df = production_df.merge(dates, left_on='Year', right_on = 'date', how='left')
production_df.head()
```

```python
# Add location id
# Import location table
locations = pd.read_csv("location_table.csv", sep = "|")
locations = locations[['locationID', 'name']]
locations.head()
```

```python
production_df = production_df.merge(locations, left_on='Area', right_on = 'name', how='left')
production_df.head()
```

```python
# Change column name to value
production_df['value'] = production_df['Value']
```

```python
# Filter only needed columns
production_df = production_df[['factID', 'measureID', 'dateID', 'locationID', 'value']]
production_df.head()
```

```python
# Export to parquet
production_df.to_parquet('production_table_20200630.parquet',
              compression='uncompressed', index = False)
```

```python
# Export to csv
production_df.to_csv('production_table_20200630.csv', sep='|', encoding='utf-8', index = False)
```
