```python
# Imports
from pathlib import Path
import pandas as pd
import geopandas as gpd
from fastparquet import ParquetFile
```

```python
# Import shapefiles

# Countries
shp0_Kenya = gpd.read_file("data/gadm36_KEN_0.shp")
shp0_Ethiopia = gpd.read_file("data/gadm36_ETH_0.shp")
shp0_Somalia = gpd.read_file("data/gadm36_SOM_0.shp")

# Regions1
shp1_Kenya = gpd.read_file("data/gadm36_KEN_1.shp")
shp1_Ethiopia = gpd.read_file("data/gadm36_ETH_1.shp")
shp1_Somalia = gpd.read_file("data/gadm36_SOM_1.shp")

# Regions2
shp2_Kenya = gpd.read_file("data/gadm36_KEN_2.shp")
shp2_Ethiopia = gpd.read_file("data/gadm36_ETH_2.shp")
shp2_Somalia = gpd.read_file("data/gadm36_SOM_2.shp")

# Regions3
shp3_Kenya = gpd.read_file("data/gadm36_KEN_3.shp")
shp3_Ethiopia = gpd.read_file("data/gadm36_ETH_3.shp")
#shp3_Somalia = gpd.read_file("data/gadm36_SOM_3.shp") #doesn't exist
```

```python
gdf_countries_list = [shp0_Kenya, shp0_Ethiopia, shp0_Somalia]
countries = gpd.GeoDataFrame(pd.concat(gdf_countries_list, ignore_index=True))
countries["locationID"] = countries["GID_0"]
countries["name"] = countries["NAME_0"]
countries["hierarchy"] = 0
countries["type"] = "Country"
countries = countries[["locationID", "name", "hierarchy", "type", "GID_0", "NAME_0"]]
countries.head()
```

```python
gdf_region1_list = [shp1_Kenya, shp1_Ethiopia, shp1_Somalia]
region1 = gpd.GeoDataFrame(pd.concat(gdf_region1_list, ignore_index=True))
region1["locationID"] = region1["GID_1"]
region1["name"] = region1["NAME_1"]
region1["type"] = "Region1"
region1["hierarchy"] = 1
region1 = region1[["locationID", "name", "hierarchy", "type", "GID_0", "NAME_0", "GID_1", "NAME_1"]]
print(region1.head())
```

```python
gdf_region2_list = [shp2_Kenya, shp2_Ethiopia, shp2_Somalia]
region2 = gpd.GeoDataFrame(pd.concat(gdf_region2_list, ignore_index=True))
region2["locationID"] = region2["GID_2"]
region2["name"] = region2["NAME_2"]
region2["type"] = "Region2"
region2["hierarchy"] = 2
region2 = region2[["locationID", "name", "hierarchy", "type", "GID_0", "NAME_0", "GID_1", "NAME_1", "GID_2", "NAME_2"]]
print(region2.head())
```

```python
gdf_region3_list = [shp3_Kenya, shp3_Ethiopia]
region3 = gpd.GeoDataFrame(pd.concat(gdf_region3_list, ignore_index=True))
region3["locationID"] = region3["GID_3"]
region3["name"] = region3["NAME_3"]
region3["hierarchy"] = 3
region3["type"] = "Region3"
region3 = region3[["locationID", "name", "hierarchy", "type", "GID_0", "NAME_0", "GID_1", "NAME_1", "GID_2", "NAME_2", "GID_3", "NAME_3"]]
print(region3.head())
```

```python
gdf_all_list = [countries, region1, region2, region3]
gdf_all = gpd.GeoDataFrame(pd.concat(gdf_all_list, ignore_index=True))
gdf_all = gdf_all[['locationID', 'name', 'type', 'hierarchy', 'GID_0', 'GID_1', 'GID_2', 'GID_3', 'NAME_0', 'NAME_1', 'NAME_2',
       'NAME_3']]
gdf_all.head(10)
```

```python
# Export to parquet
gdf_all.to_parquet('location_table.parquet',
              compression='uncompressed')
```

```python
# Open parquet file
result = pd.read_parquet('location_table.parquet', engine='fastparquet')
result.head()
```
