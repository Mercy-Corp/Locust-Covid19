```python
# Imports
from pathlib import Path
import pandas as pd
import geopandas as gpd
from fastparquet import ParquetFile
```

```python
# Import shapefiles

# Countries
shp0_Kenya = gpd.read_file("data/gadm36_KEN_0.shp")
shp0_Ethiopia = gpd.read_file("data/gadm36_ETH_0.shp")
shp0_Somalia = gpd.read_file("data/gadm36_SOM_0.shp")

# Regions1
shp1_Kenya = gpd.read_file("data/gadm36_KEN_1.shp")
shp1_Ethiopia = gpd.read_file("data/gadm36_ETH_1.shp")
shp1_Somalia = gpd.read_file("data/gadm36_SOM_1.shp")

# Regions2
shp2_Kenya = gpd.read_file("data/gadm36_KEN_2.shp")
shp2_Ethiopia = gpd.read_file("data/gadm36_ETH_2.shp")
shp2_Somalia = gpd.read_file("data/gadm36_SOM_2.shp")

# Regions3
shp3_Kenya = gpd.read_file("data/gadm36_KEN_3.shp")
shp3_Ethiopia = gpd.read_file("data/gadm36_ETH_3.shp")
#shp3_Somalia = gpd.read_file("data/gadm36_SOM_3.shp") #doesn't exist
```

```python
gdf_countries_list = [shp0_Kenya, shp0_Ethiopia, shp0_Somalia]
columns = ["locationID", "name", "geometry"]
countries = gpd.GeoDataFrame(pd.concat(gdf_countries_list, ignore_index=True))
countries.columns = columns
countries = countries[["locationID", "geometry"]]
countries.head()
```

```python
gdf_regions1_list = [shp1_Kenya, shp1_Ethiopia, shp1_Somalia]
columns = ["locationID", "name", "type", "geometry"]
regions1 = gpd.GeoDataFrame(pd.concat(gdf_regions1_list, ignore_index=True))
regions1 = regions1[["GID_1", "NAME_1", "TYPE_1", "geometry"]]
regions1.columns = columns
regions1 = regions1[["locationID","geometry"]]
print(regions1.shape)
regions1.head()
```

```python
print(shp2_Kenya.shape)
shp2_Kenya.head(2)
```

```python
gdf_regions2_list = [shp2_Kenya, shp2_Ethiopia, shp2_Somalia]
columns = ["locationID", "name", "type", "geometry"]
regions2 = gpd.GeoDataFrame(pd.concat(gdf_regions2_list, ignore_index=True))
regions2 = regions2[["GID_2", "NAME_2", "TYPE_2", "geometry"]]
regions2.columns = columns
regions2 = regions2[["locationID", "geometry"]]
print(regions2.shape)
regions2.head()
```

```python
print(shp3_Kenya.shape)
shp3_Kenya.head(2)
```

```python
gdf_regions3_list = [shp3_Kenya, shp3_Ethiopia]
columns = ["locationID", "name", "type", "geometry"]
regions3 = gpd.GeoDataFrame(pd.concat(gdf_regions3_list, ignore_index=True))
regions3 = regions3[["GID_3", "NAME_3", "TYPE_3", "geometry"]]
regions3.columns = columns
regions3 = regions3[["locationID", "geometry"]]
print(regions3.shape)
regions3.head()
```

```python
gdf_all_list = [countries, regions1, regions2, regions3]
gdf_all = gpd.GeoDataFrame(pd.concat(gdf_all_list, ignore_index=True))
gdf_all.head(10)
```

```python
# Export to parquet
gdf_all.to_parquet('shapefile_table.parquet',
              compression='uncompressed')
```

```python
gdf_all.to_file("shapefile_table.csv", sep='|', encoding='utf-8')
#gdf_all.to_file("shapefile_table.json", driver="GeoJSON")
```
